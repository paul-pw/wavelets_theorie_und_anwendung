```python
import pywt
import matplotlib.pyplot as plt
```

```python
pywt.wavelist()
```

```python
[phi, psi, x] = pywt.Wavelet('db2').wavefun(level=10)
plt.plot(x,psi)
plt.show()
```

```python
[phi, psi, x] = pywt.Wavelet('haar').wavefun(level=10)
plt.plot(x,psi)
plt.show()
```

```python
[psi, x] = pywt.ContinuousWavelet('mexh').wavefun(level=10)
plt.plot(x,psi)
plt.show()
```

```python
[psi, x] = pywt.ContinuousWavelet('morl').wavefun(level=10)
plt.plot(x,psi)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
fig.set_figwidth(10)
fig.set_figheight(8)

[phi, psi, x] = pywt.Wavelet('haar').wavefun(level=10)
ax[0,0].plot(x,psi)
ax[0,0].set_title('Haar',loc='center', fontsize=15)

[phi, psi, x] = pywt.Wavelet('db2').wavefun(level=10)
ax[0,1].plot(x,psi)
ax[0,1].set_title('Daubechies',loc='center', fontsize=15)

[psi, x] = pywt.ContinuousWavelet('mexh').wavefun(level=10)
ax[1,0].plot(x,psi)
ax[1,0].set_title('Mexican Hat',loc='center', fontsize=15)

[psi, x] = pywt.ContinuousWavelet('morl').wavefun(level=10)
ax[1,1].set_title('Morlet',loc='center', fontsize=15)
ax[1,1].plot(x,psi)

plt.tight_layout()
plt.savefig("assets/wavelets.png", dpi = 300)
plt.show()
```
